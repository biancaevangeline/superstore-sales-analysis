```python
import pandas as pd

df = pd.read_csv("data/train.csv", index_col = "Row ID")
df
```

```python
df.info()
```

```python
#Changing "Ship Date" and "Ship Date" column to datetime

df["Order Date"] = pd.to_datetime(df["Order Date"], format = "%d/%m/%Y")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format = "%d/%m/%Y")
```

```python
#Getting rid of duplicates

df = df.drop_duplicates()
```

```python
#Removing NaN values from the "Postal Code" column

df = df.dropna(subset = "Postal Code")
```

```python
#Changing "Postal Code" column from a float to string

df["Postal Code"] = df["Postal Code"].astype(str)
```

```python
#Save 1_data_cleaning file 

df.to_csv("data/superstore_cleaned.csv", index = True)
```
