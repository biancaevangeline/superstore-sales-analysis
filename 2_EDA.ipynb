```python
import pandas as pd

df = pd.read_csv("data/superstore_cleaned.csv", index_col = "Row ID")

#Re-apply dtypes
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df["Postal Code"] = df["Postal Code"].astype(str)

df
```

```python
#Making a new column for year, month, and month no
df["Order Date Year"] = df["Order Date"].dt.year
df["Order Date Month"] = df["Order Date"].dt.strftime("%b")
df["Order Date Month No"] = df["Order Date"].dt.month
df
```

```python
df.describe()
```

```python
#Sorting the month using "Order Date Month No" column, then grouping it by year using for loop

df = df.sort_values("Order Date Month No")

years = [2015, 2016, 2017, 2018]
dfs = {}

for year in years:
    dfs[year] = df[df["Order Date Year"] == year].copy()

df
```

```python
#Median of Sales by Year 

df_year_sales = df.pivot_table(
    index = ["Order Date Month No", "Order Date Month"],
    columns = "Order Date Year",
    values = "Sales",
    aggfunc = "median"
).sort_index().droplevel("Order Date Month No")


df_year_sales
```

```python
#Total sales by Category from 2015 to 2018 

df_cat_year = df.pivot_table(
    index = "Order Date Year",
    columns = "Category",
    values = "Sales",
    aggfunc = "sum"
)

df_cat_year
```

```python
#Top 10 Most Selling Sub-Category by year 
years = [2015, 2016, 2017, 2018]
df_subcat = {}

for year in years:
    df_subcat[year] = dfs[year].groupby("Sub-Category")["Sales"].sum().sort_values(ascending = False).head(10)

#testing the loop
df_subcat[2015]
```

```python
#Sales by Ship mode per segment
df_shipmode_segment = df.pivot_table(
    index = "Ship Mode",
    columns = "Segment",
    values = "Sales",
    aggfunc = "median"
)

df_shipmode_segment
```

```python
#Top 5 states with most orders by region 

df_east = df[df["Region"] == "East"].groupby("State").size().sort_values(ascending = False).head()
df_west = df[df["Region"] == "West"].groupby("State").size().sort_values(ascending = False).head()
df_central = df[df["Region"] == "Central"].groupby("State").size().sort_values(ascending = False).head()
df_south = df[df["Region"] == "South"].groupby("State").size().sort_values(ascending = False).head()

```

```python
#Sales by Segment from 2015 - 2018

years = [2015, 2016, 2017, 2018]
df_sales_segment = {}

for year in years:
    df_sales_segment[year] = dfs[year].groupby("Segment")["Sales"].sum()

#testing the loop
df_sales_segment[2015]
```
